```python
import numpy as np
import pandas as pd
```

```python
rolls = np.random.choice(range(1,7), size=(1_000_000, 3))
rolls
```

```python
rolls = pd.DataFrame(rolls)
```

```python
rolls
```

```python
rolls.rename(columns={0: "die1", 1: "die2", 2: "die3"}, inplace=True)
```

```python
rolls
```

```python
lucky_number = 5
```

```python
rolls["first_matches"] = rolls["die1"] == lucky_number
rolls["second_matches"] = rolls["die2"] == lucky_number
rolls["third_matches"] = rolls["die3"] == lucky_number
```

```python
rolls.astype(int)
```

```python
rolls["number_of_matches"] = rolls["first_matches"] + rolls["second_matches"] + rolls["third_matches"]
```
