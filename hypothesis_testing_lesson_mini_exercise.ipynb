2. Write the code necessary to simulate taking the average of 6 dice rolls 10,000 times. Store each of the 10,000 averages in a list.

```python
n_simulations = 10_000
n_trials = 6

df = pd.DataFrame(np.random.choice(range(1,7), size=(n_simulations, n_trials)))
df
```

```python
df["mean"] = df.mean(axis=1)
df
```

```python
roll_means = []

for i in range(10_000):
    roll_mean = np.random.choice(range(1,7)).mean(axis=1)
    roll_means.append(roll_mean)
```
